# Figures
Figures for visualising performance comparison between the different Bayesian Networks created for the project: Investigating Probabilistic Performance Prediction of MapReduce Applications using Bayesian Networks

#### Import dependecies

```python
import pandas as pd
import numpy as np
import networkx as nx
import matplotlib.pyplot as plt
```

#### Import data

```python
df = pd.read_csv('Model-metrics.csv', header=0, index_col=False)

df.head()
```

```python
fig_size = (8, 5)
```

```python
ax = df.plot.bar(x='Model', y='F1', rot=0, figsize=fig_size, title='F1 score comparison by model')
```

```python
ax = df.plot.bar(x='Model', y='Accuracy', rot=0, figsize=fig_size, title='Accuracy score comparison by model')

for index, value in enumerate(list(df["Accuracy"])):
    plt.text(index, value, str(value))
plt.show()
```

```python
ax = df.plot.bar(x='Model', y='Precision', rot=0, figsize=fig_size, title='Precision score comparison by model')
```

```python
ax = df.plot.bar(x='Model', y='Recall', rot=0, figsize=fig_size, title='Recall score comparison by model')
```

```python
ax = df.plot.bar(x='Model', y='Log likelihood score', rot=0, figsize=fig_size, title='Log likelihood score comparison by model')
```

```python
ax = df.plot.bar(x='Model', y='Structure score', rot=0, figsize=fig_size, title='Structure score comparison by model')
```

```python
ax = df.plot.bar(rot=0, figsize=fig_size)
```

```python
import seaborn as sns
ax = df.plot.bar(x='Model', y='TT prediction accuracy', rot=0, figsize=fig_size, title='Transaction time prediction accuracy as percentage (%)')

plt.figure(figsize=(8, 6))
plots = sns.barplot(x="Model", y="TT prediction accuracy", data=df)

# Setting the x-acis label and its size
plt.xlabel("Model", size=15)
 
# Setting the y-axis label and its size
plt.ylabel("Accuracy (%)", size=15)
 
# Finallt plotting the graph
plt.show()
```
